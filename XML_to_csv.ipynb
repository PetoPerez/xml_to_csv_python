```python
#importamos las librerías a utilizar
import xml.etree.ElementTree as Xet
import pandas as pd
```

```python
#creamos dos listas columnas y filas a utilizar
cols = ["Año", "Número_de_Subasta", "Tipo_de_Instrumento", "Clave_de_Emisión", "Plazo", "Fecha_de_Subasta", "Fecha_de_Colocacion",
       "Fecha_de_Vencimiento", "Sobretasa_Ponderada", "Precio_Ponderado", "Monto_Convocado", "Monto_Solicitado", "Monto_Colocado"]
rows = []
```

```python
#leemos el archivo XML y mediante un for descomponemos la información
xmlparse = Xet.parse('Data.xml')
root = xmlparse.getroot()
for i in root:
    Año = i.find("Año").text
    Número_de_Subasta = i.find("Número_de_Subasta").text
    Tipo_de_Instrumento = i.find("Tipo_de_Instrumento").text
    Clave_de_Emisión= i.find("Clave_de_Emisión").text
    Plazo = i.find("Plazo").text
    Fecha_de_Subasta = i.find("Fecha_de_Subasta").text
    Fecha_de_Vencimiento = i.find("Fecha_de_Vencimiento").text
    Sobretasa_Ponderada = i.find("Sobretasa_Ponderada").text
    Precio_Ponderado = i.find("Precio_Ponderado").text
    Monto_Convocado = i.find("Monto_Convocado").text
    Monto_Solicitado = i.find("Monto_Solicitado").text
    Monto_Colocado = i.find("Monto_Colocado").text
    rows.append({"Año": Año,
                 "Número_de_Subasta": Número_de_Subasta,
                 "Tipo_de_Instrumento": Tipo_de_Instrumento,
                 "Clave_de_Emisión": Clave_de_Emisión,
                 "Plazo": Plazo,
                 "Fecha_de_Subasta": Fecha_de_Subasta,
                 "Fecha_de_Vencimiento": Fecha_de_Vencimiento,
                 "Sobretasa_Ponderada": Sobretasa_Ponderada,
                 "Precio_Ponderado": Precio_Ponderado,
                 "Monto_Convocado": Monto_Convocado,
                 "Monto_Solicitado": Monto_Solicitado,
                 "Monto_Colocado": Monto_Colocado           
                 })
```

```python
#creamos el data frame desde las listas que creamos y ya les agregamos la data
df = pd.DataFrame(rows, columns=cols)
```
